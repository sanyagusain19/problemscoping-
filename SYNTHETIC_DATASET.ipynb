```python
import numpy as np
import pandas as pd
```

<div style="
background-color:#FFF8E7;
padding:6px;
">
<h1 style="color:#74695A;">
Generating data:
</h1>
</div>

```python
np.random.seed(6)
n_data=10000
data={
    "customer_id": [f"{i:05d}"for i in range(1,n_data+1)],
    "Tenure": np.random.randint (0,12,n_data),
    "City_tier": np.random.choice([1,2,3],n_data,p=[0.3,0.5,0.2]),
    "Gender": np.random.choice(["Male","Female"],n_data,p=[0.5,0.5]),
    "hour_spend_on_app": np.random.poisson(5,n_data),
    "Number_devices_registered": np.random.randint(0,5,n_data),
    "prefered_ordered_cat": np.random.choice(["electronics","assesories","furniture"],n_data,p=[0.4,0.4,0.2]),
    "satisfaction_score": np.clip(np.random.normal(3,2,n_data),0,5).round(1),
    "No_of_complaints": np.random.randint(0,4,n_data),
    "coupon_used" : np.random.choice([0,1],n_data,p=[0.6,0.4]),
    "order_count" : np.clip(np.random.normal(5,3,n_data),0,10).round(1),
    "DaySinceLastOrder" : np.clip(np.random.normal(5,2,n_data) , 0, 10).round(1),
    "cashback" : np.random.choice([0,1],n_data,p=[0.3,0.7])
}
```

<div style="
background-color:#FFF8E7;

```python
df = pd.DataFrame(data)
print(df)
```
